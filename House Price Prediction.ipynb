```python
train,valid,test= np.split(df.sample(frac=1),[int(0.6*len(df)),int(0.8*len(df))])
```

```python
def standard_scaler(dataframe, y_label, x_labels=None):
    dataframe = copy.deepcopy(dataframe)

    if x_labels is None:
        X = dataframe[[c for c in dataframe.columns if c != y_label]].values
    else:
        if len(x_labels) == 1:
            X = dataframe[x_labels[0]].values.reshape(-1, 1)
        else:
            X = dataframe[x_labels].values

    Y = dataframe[y_label].values.reshape(-1, 1)
    data = np.hstack((X, Y))
    return data, X, Y
```

```python
x_labels = [col for col in df.columns if col != 'medv']
train, x_train,y_train= standard_scaler(train,'medv',x_labels=x_labels)
valid, x_valid,y_valid= standard_scaler(valid,'medv',x_labels=x_labels)
test,x_test,y_test=standard_scaler(test,'medv',x_labels=x_labels)
```

```python
lr_model= LinearRegression()
lr_model.fit(x_train,y_train)
```

```python
predict_y= lr_model.predict(x_test)
```

```python
error = abs(predict_y[1] - y_test[1])
print(f'Actual amount: ${y_test[1]*1000}')
print(f'Prediction amount: ${predict_y[1]*1000}')
print(f"Absolute error: {error[0]}")
print(f'Actual predicted price: ${(predict_y[1]*1000)-(error[0]*1000)}')
```
